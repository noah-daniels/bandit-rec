### Sample Average
$\overline{\mu}=\dfrac{c}{n}$
+ $c$ clicks
+ $n$ impressions

### Froomle
$\overline{\mu_1}=\dfrac{c+b \cdot b_{ctr}}{n+b}$
+ $b$ boost denominator
+ $b_{ctr}$ boost ctr

### Dynamic Froomle
$\overline{\mu_2}=(1-\lambda)\overline{\mu} + \lambda b_{ctr}$

$\lambda = \dfrac{\max\{b-n, 0\}}{b} $

+ $b_{ctr} = f_1 \cdot CTR_{top}$
+ $b = \dfrac{f_2}{b_{ctr}}$
